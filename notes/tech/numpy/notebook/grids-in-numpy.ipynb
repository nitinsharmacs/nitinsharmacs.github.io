# Grids in numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
a = np.array([3, 1, 2, 5, 6, 7])
b = np.array(a[::-1])
x, y = np.meshgrid(a, b)
```

```python
x, y
```

```python
plt.plot(x, y, marker='.', color='k', linestyle='none')
plt.xticks(a)
plt.yticks(a)
plt.show()
```

## Remarks

1) Using meshgrid, we can create coordinate metrices `x`, `y`.
2) Meshgrid allows to make all possible combinations for given arrays elements.

## Plotting two dimension functions

### Sine function

```python
def sin(x, y):
    return np.sin(x) + np.sin(y)
```

```python
z = sin(x, y)
z
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(z, origin='lower')
plt.colorbar()
plt.locator_params(axis='y', nbins=4)
plt.locator_params(axis='x', nbins=4)
plt.show()
```

```python
seq = np.linspace(-10, 10, 100)
x, y = np.meshgrid(seq, seq)
```

```python
z = sin(x, y)
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(z)
plt.colorbar()
plt.locator_params(axis='x', nbins=4)
plt.locator_params(axis='y', nbins=4)
plt.show()
```

### Another function

```python
z = x ** 2 + (y - x) * 2
```

```python
plt.imshow(z)
plt.colorbar()
plt.locator_params('x', nbins=4)
plt.locator_params('y', nbins=4)
plt.show()
```
